```python
print(s.info())
```

```python
n_val = s.isnull().sum()
print("Null Values:\n", n_val)
```

```python
s['Mileage'] = s['Mileage'].str.extract('(\d+\.\d+)')
s['Mileage'] = s['Mileage'].astype(float)
s['Mileage'].fillna(s['Mileage'].median(), inplace=True)
s
```

```python
s['Engine'] = s['Engine'].str.extract('(\d+)')
s['Engine'] = s['Engine'].astype(float)

s['Power'] = s['Power'].str.extract('(\d+\.\d+)')
s['Power'] = s['Power'].astype(float)

s['Engine'].fillna(s['Engine'].median(), inplace=True)

s['Power'].fillna(s['Power'].median(), inplace=True)

s['Seats'].fillna(s['Seats'].median(), inplace=True)
n_val= s.isnull().sum()
print("Null Values:\n",n_val)
```

```python
s.drop('New_Price',axis=1, inplace=True)
```
