```python
d3=pd.DataFrame(d1)
```

```python
# (b) Remove the units from Mileage, Engine, Power and New_Price and converting them into float
```

```python
d3['Mileage'] = d3['Mileage'].str.extract('(\d+\.\d+)').astype(float)

d3['Engine'] = d3['Engine'].str.replace(' CC', '').astype(float)

d3['Power'] = d3['Power'].str.extract('(\d+\.\d+)').astype(float)

d3['New_Price'] = d3['New_Price'].str.extract('(\d+\.\d+)').astype(float)

d3.head()
```

```python
#(a) Filling in missing values im Mileage, Engine, Power,Seats
```

```python
d3['Mileage'].fillna(d3['Mileage'].median(),inplace=True)
d3['Engine'].fillna(d3['Engine'].median(),inplace=True)
d3['Power'].fillna(d3['Power'].median(),inplace=True)

d3['Seats'].fillna(d3['Seats'].mode()[0],inplace=True)

d3.drop('New_Price', axis=1, inplace=True)#axis=1 indicates to drop a column
```

```python
output_csv_file = 'preprocessed_data.csv'
d3.to_csv(output_csv_file, index=False)
```
